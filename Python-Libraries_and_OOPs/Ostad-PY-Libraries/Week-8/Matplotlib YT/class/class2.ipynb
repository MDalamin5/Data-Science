```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2
print(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r-*')
plt.show()
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('This is X-axis Title')
plt.ylabel('This is Y-axis Title')
plt.title('Figure Ploat')
plt.xticks(range(0, 18, 3))
plt.show()
```
