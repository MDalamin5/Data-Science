```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('job-data.csv')
df.head(3)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop(columns=['joining_date'], inplace=True)
df.head(2)
```

```python
x = df.drop('Salary', axis=1)
```

```python
x
```

```python
y = df.Salary
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_x = LabelEncoder()
x = x.apply(le_x.fit_transform)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest =train_test_split(x, y, train_size=.8, random_state=3)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.tree import DecisionTreeClassifier
dect = DecisionTreeClassifier()
dect.fit(xtrain, ytrain)
```

```python
dect.predict(xtest)
```

```python
dect.score(xtest, ytest)
```
