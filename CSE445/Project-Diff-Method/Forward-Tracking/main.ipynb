```python
import pandas as pd
```

```python
df = pd.read_csv('job-data.csv')
df.head(4)
```

```python
col_1_df = df.drop(columns=['joining_date', 'job_title', 'Overall_Satisfication', 'Working_model', 'Considering_Job_Switching_in_Future', 'Main_Reasons'], axis=1)
col_1_df
```

```python
X = col_1_df.drop('Salary', axis=1)
X
```

```python
y = col_1_df.Salary
y.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
le_en = LabelEncoder()
```

```python
X = X.apply(le_en.fit_transform)
X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
my_model = LinearRegression()
my_model.fit(X_train, y_train)
```

```python
my_model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
result = cross_val_score(DecisionTreeClassifier(), X, y, cv=5)
result
```

```python
from sklearn.neighbors import KNeighborsClassifier
result = cross_val_score(KNeighborsClassifier(n_neighbors=3), X, y, cv=5)
result
```
